```python
red_vals = []
for _ in range(10000):
    rolls = np.random.choice(bag, size = 4, replace = True)
    count_r = sum(rolls == 'R')
    red_vals.append(count_r)
```

```python
len(red_vals)
```

```python
pd.Series(red_vals).value_counts()
```

```python
pd.value_counts(red_vals, normalize=True)
```
